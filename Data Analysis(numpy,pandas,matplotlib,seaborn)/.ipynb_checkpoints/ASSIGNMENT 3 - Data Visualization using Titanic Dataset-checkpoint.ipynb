```python
df1=pd.read_csv('titanic/train.csv')
```

```python
print(df1.columns)
df1.head()
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1['Age']=df1['Age'].fillna(np.mean(df1['Age']))
```

```python
df1.isna().sum()
```

```python
df1.drop(columns='Cabin',inplace=True)
```

```python
#drop Null values
df1=df1.dropna()
df1.isna().sum()
```

```python
#drop duplicates
df1=df1.drop_duplicates()
```

```python
df1.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df1)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(df1['PassengerId'],df1['Age'],hue=df1['Sex'])
plt.title("PassengerId Vs Age")
plt.xlabel("PassengerId")
plt.ylabel("Age")
plt.show()
```

```python
#Box plot 
sns.barplot(df1['PassengerId'],df1['SibSp'],hue=df1['Sex'],data=df1)
plt.show()
```
