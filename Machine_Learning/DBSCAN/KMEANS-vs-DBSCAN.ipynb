# KMEANS vs DBSCAN clustering

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(blobs)
```

```python
model.labels_
```

# Label Discovery

```python
def display_categories(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
display_categories(model,blobs)
```

```python
cir_model = KMeans(n_clusters=2)
cir_model.fit_transform(circles)
```

```python
display_categories(cir_model,circles)
```

```python
moon_model = KMeans(n_clusters=2)
moon_model.fit_transform(moons)
```

```python
display_categories(moon_model,moons)
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
```

```python
help(DBSCAN)
```

```python
dbscan = DBSCAN(eps=2)
```

```python
dbscan.fit(blobs)
```

```python
display_categories(dbscan,blobs)
```

```python
moon_dbs = DBSCAN(eps=0.15)
moon_dbs.fit(moons)
```

```python
display_categories(moon_dbs,moons)
```

```python
cir = DBSCAN(eps=.4)
display_categories(cir,circles)
```
