## Feature Selection

```python
X = df[['floors','waterfront','lat','bedrooms','sqft_basement','view','bathrooms','sqft_living15','sqft_above','grade','sqft_living']]
y = df['price']
```

# #train-test-split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101,test_size=0.3)
```

# #feature Standardlization

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train_scalr = std.fit_transform(X_train)
X_test_sclar = std.transform(X_test)
```

# #LogesticRegression model build

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scalr,y_train)
```

```python
y_pred = model.predict(X_test_sclar)
```

# #Model Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
# #Best Degree value(K)
r2_Score = []
for k in range(1,10):
    poly_model = PolynomialFeatures(degree=k,include_bias=False)
    X_feature = poly_model.fit_transform(X)
    X_train,X_test,y_train,y_test = train_test_split(X_feature,y,random_state=101,test_size=0.3)
    X_train_Sclar = std.fit_transform(X_train)
    X_test_Sclar = std.transform(X_test)
    model.fit(X_train_Sclar,y_train)
    y_pred = model.predict(X_test_Sclar)
    r2_Score.append(r2_score(y_test,y_pred))
    
```
