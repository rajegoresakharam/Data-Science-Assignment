```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Datasets/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sclar = StandardScaler()
```

```python
X_sclar = sclar.fit_transform(df)
```

```python
X_sclar
```

## Priciple Components Analysis

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components=2)
```

```python
p_compts = pca_model.fit_transform(X_sclar)
```

```python
pca_model.explained_variance_
```

```python
pca_model.explained_variance_ratio_
```

```python
pca_model.get_covariance()
```

```python
pca_model.components_
```

```python
plt.figure(figsize=(6,4),dpi=200)
sns.scatterplot(p_compts[:,0],p_compts[:,1])
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
print(label_data['DESCR'])
```

```python
label = label_data['target']
```

```python
plt.figure(figsize=(6,4),dpi=200)
sns.scatterplot(p_compts[:,0],p_compts[:,1],hue=label)
plt.show()
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PCA1','PCA2'],columns=df.columns)
df_comp
```

# #best n_components value

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(X_sclar)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(4,3),dpi=200)
sns.lineplot(range(1,30),explained_variance)
plt.show()
```
